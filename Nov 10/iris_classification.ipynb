```python
iris = pd.read_csv('iris.csv')
iris.columns = ['petal_length', 'petal_width', 'calyx_length', 'calyx_width', 'class']
print(iris.columns)
iris.head()
```

```python
# onehot-encoding
incoding = pd.get_dummies(iris)
print(incoding.columns)
incoding.head()
```

```python
# independent value (inde_val), dependent value (de_val)
inde_val = incoding[['petal_length', 'petal_width', 'calyx_length', 'calyx_width']]
de_val = incoding[['class_setosa', 'class_versicolor','class_virginica']]
print(inde_val.shape, de_val.shape)
```

```python
# model
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3,activation='softmax')(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',
             metrics = 'accuracy')
```

```python
model.fit(inde_val, de_val, epochs=200)
```

```python
print(model.predict(inde_val[0:5]))
print(de_val[0:5])
```

```python
print(model.predict(inde_val[-5:]))
print(de_val[-5:])
```
